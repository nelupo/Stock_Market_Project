```python
# stock close price - graph 

import numpy as np
import pandas_datareader as pdr
import datetime
from datetime import date
import yfinance as yf
import plotly.express as px
import pandas as pd
import finplot as fplt
import matplotlib.pyplot as plt
import math
today = date.today()

# CHANGE STOCK:
stock = 'CELZ'
# can add more than 1

stockData = yf.download(stock)
df = yf.download(stock)

stock_price_chart = px.area(stockData['Close'],
                            title=stock + ' ' + '- Daily Close Price',
                            color_discrete_map={'Close': 'green'},
                            width=800, height=800)
stock_price_chart.show()
```
